```python
from ultralytics import YOLO
import os
from google.colab import files
```

```python
from ultralytics import YOLO
# Also used yolov9c to train another model
model = YOLO('yolov9s.yaml')
```

```python
#Training code
results = model.train(data='data.yaml', epochs=40,batch=16, lr0=0.01, optimizer='AdamW', weight_decay=0.0004)
```

```python
#Perform Predictions on a sample image

model_path = '/train5/weights/best.pt' #path to the weights file

if not os.path.exists(model_path):
    print(f"Error: Model file not found at {model_path}")
else:
    model = YOLO(model_path)
    uploaded = files.upload()
    for fn in uploaded.keys():
        img_path = fn
        results = model(img_path,conf=0.4)
        results[0].show() # Displays the image with bounding boxes
```
